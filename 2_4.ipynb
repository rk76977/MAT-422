## **2.4.1. MLE for random samples**

Maximum likelyhood esimation estimates parameters from samples by considering the the parameters $θ$ so that the pdf $f(x_1, x_2, \dots x_n | θ)$ is maximized. This gives the $θ$ so that the probability of having observed the sample is largest.

One interesting application of MLE is determining the $p$ so that a sample of $x_1 \dots x_n$ of the Bernoulli distribution was most likely to occur given $p$. If there were $k$ 1s observed and $n-k$ 0s, then the probability of $x_1 \dots x_n$ is $p^k(1-p)^{n-k}$. Finding the $p$ that minimizes this is $p =\frac{k}{n}$. Intuitively, given $k$ 1s out of $n$, we expect the probabity associated with the bernoulli distribution to be $\frac{k}{n}$.

## **2.4.2. Linear regression**

### Applying linear regression

One interesting example of regression is approximating a non-polynomial function as a polynomial. If we have a function $f(x)$ that is differentiable but might not have an easy computable $k$th derivative for a taylor series, we can approximate $f(x)$ as $a_0+a_1x + \dots a_nx^n$.

In this example let $f(x) = x\log(x)$. We can approximate $f(x)$ as a linear combination of $a_0 , \dots a_5x^5$

We find the coefficients using regression and take the matrix as values of each $x^k$ and the $y$ matrix as the value of $f$ for each $x$

```python
import numpy as np
import matplotlib.pyplot as plt
import math
def f(x):
  return x*math.log(x)


data = np.empty((100,5))
Y=np.empty((100,1))
for i in range(data.shape[0]):
  x = np.random.uniform(0, 10)
  row = np.array([x**i for i in range(0,5)])
  data[i:] = row

  y = f(x)
  Y[i:]=[y]


coefficients = (np.linalg.inv(data.T @ data) @ data.T) @ Y

print(f'g(n): {coefficients[0]} +  {coefficients[1]}n + {coefficients[2]}n^2 + {coefficients[3]}n^3 + {coefficients[4]}n^4')
```
