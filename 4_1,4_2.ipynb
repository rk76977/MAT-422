## 4.1.1 Graph Models

For a graph $G(V,E)$, the adjacency matrix is the $|v|$-by-$|v|$ matrix of the form $A_{ij} = E_{ij}$, where $E_{ij}=0$ if node $i$ is not adjacent to node $j$.

Here is an example of a graph

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_weighted_edges_from([(1,2,1),(1,3,1),(2,3,1),(3,2,1),(2,1,1),(3,1,1),(3,4,10),(4,3,10)])

positions = nx.spring_layout(G)
nx.draw(G, positions, with_labels=True,)

nx.draw_networkx_edge_labels(G, positions, edge_labels=nx.get_edge_attributes(G, 'weight'))

plt.show()
```

The adjacency matrix is $A=\begin{pmatrix}
0&1&1&0\\
1&0&1&0\\
1&1&0&10\\
0&0&10&0
\end{pmatrix}$


Using software, the largest eigenvalue is about 10.1104, and gives a corresponding normalized eigenvector of
$\begin{pmatrix} 0.0775696\\ 0.0775696\\ 0.706687\\ 0.69897 \end{pmatrix}$.
I learned that interestingly, the largest entries of this eigenvector correspond to how central a node is. Thus, node 3 is the most central, and despite node 4 only being connected to node 3, it is the second most central because its edge to node 3 has weight 10.
