## 3.7.1 Mathematical Formulation

Networks are collections of neurons, where each neuron is a linear function that applies an activation. This can be written as $A(\mathbf{W_i}x_i+b_i)$ for each layer. One interesting note is that when no activation is applied, the network can be written linearly as $\mathbf{W}x+b$ because it is the composition of $\mathbf{W_i}x_i+b_i)$

## 3.7.2 Activation Functions

Activations are (usually nonlinear) functions applied on neurons. One interesting property of any bounded nonconstant continuous activation function is that there exists a network that approximates any function as precisely as desired.

Here is a function of relu's that can be used to build an arbitrary precise approximation for any function on $\mathbb{R}$

```python
import numpy as np
import matplotlib.pyplot as plt


def relu(x):
    return np.maximum(x,0)

def g(x):
  return relu(relu(x+1)-2*relu(x)) +1

G = np.vectorize(g)

x = np.linspace(-2, 2, 100)
y = G(x)

# Plot the function
plt.plot(x, y)
plt.legend()
plt.grid(True)
plt.show()
```
